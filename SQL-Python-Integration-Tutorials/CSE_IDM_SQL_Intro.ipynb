## Using Wildcards

* A wildcard character is used to substitute one or more characters in a string.
  
* Wildcard characters are used with the LIKE operator.
  
* The LIKE operator is used in a WHERE clause to search for a specified pattern in a column.

#### List of wildcards:

| Symbol | Description|
|--------|--------------------------------------------------------------|
| %      | Represents zero or more characters                           |
| _      | Represents a single character                                |
| []     | Represents any single character within the brackets *        |
| ^      | Represents any character not in the brackets *               |
| -      | Represents any single character within the specified range * |
| {}     | Represents any escaped character **                          |

```python
# Using % wildcard: Select names starting with E

actors_df_10 = pd.read_sql("""SELECT *
                              FROM actor
                              WHERE first_name LIKE "E%";""", connection_db)  ## % represents zero or more characters
print(actors_df_10.shape)
actors_df_10
```

```python
# Using % wildcard: Selecting names starting with E and ending with N

actors_df_11 = pd.read_sql("""SELECT *
                              FROM actor
                              WHERE first_name LIKE "E%N";""", connection_db)  ## % represents zero or more characters
print(actors_df_11.shape)
actors_df_11.head()
```

```python
# Using % wildcard: Selecting names starting with E and containing L

actors_df_11 = pd.read_sql("""SELECT *
                              FROM actor
                              WHERE first_name LIKE "E%L%";""", connection_db)  ## % represents zero or more characters
print(actors_df_11.shape)
actors_df_11
```

```python
# Using _ wildcard: Select names ending with D and having 2 letters

actors_df_12 = pd.read_sql("""SELECT *
                              FROM actor
                              WHERE first_name LIKE "_D";""", connection_db)  ## _ represents a single character
print(actors_df_12.shape)
actors_df_12.head()
```

#### For `_` wildcard one has to be specific

```python
# Using _ wildcard: Select names ending with A and having 5 letters

actors_df_13 = pd.read_sql("""SELECT *
                              FROM actor
                              WHERE first_name LIKE "____A";""", connection_db)  ## _ represents a single character
print(actors_df_13.shape)
actors_df_13
```

#### Note: All DBMS does not have the same wildcards. Some of these wildcards may or may not exist in a particular DBMS. Be wary of the wildcards allowed in a DBMS.

## Sort: Order by

The `ORDER BY` command is used to sort the result set in ascending or descending order.

The `ORDER BY` command sorts the result set in ascending order by default. To sort the records in descending order, use the `DESC` keyword.

* Data displayed appears in the order of the
underlying tables.
* Updated and deleted data can change this
order.
* Sequence of retrieved data cannot be
assumed if order was not specified.
* Sorting data logically helps keep
information you want on top.
* ORDER BY clause allows user to sort data
by particular columns.

### Rules of order by:
* Takes the name of one or more columns
* Add a comma after each additional column
name
* Can sort by a column not retrieved
* Can be sorted using column position

```python
actors_df_14 = pd.read_sql("""SELECT *
                              FROM actor
                              ORDER BY first_name, last_name;""", connection_db)
print(actors_df_14.shape)
actors_df_14.head(6)
```

#### Sorting with ASC in one and DESC in other

```python
actors_df_15 = pd.read_sql("""SELECT *
                              FROM actor
                              ORDER BY first_name, last_name DESC;""", connection_db)
print(actors_df_15.shape)
actors_df_15.head(6)
```

#### Depends on the order of the column!

```python
actors_df_15 = pd.read_sql("""SELECT *
                              FROM actor
                              ORDER BY last_name DESC, first_name;""", connection_db)
print(actors_df_15.shape)
actors_df_15.head()
```

#### Use the column position appearing in the SELECT statement 

The positional index for the first column in the SELECT statement is 1, the second is 2, etc. 

```python
actors_df_15 = pd.read_sql("""SELECT *
                              FROM actor
                              ORDER BY 3 DESC, 2;""", connection_db)
print(actors_df_15.shape)
actors_df_15.head()
```

#### Can sort by a column not retrieved 

```python
actors_df_15 = pd.read_sql("""SELECT actor_id, last_update
                              FROM actor
                              ORDER BY last_name DESC, first_name;""", connection_db)
print(actors_df_15.shape)
actors_df_15.head()
```

## Aggregate Functions:

An aggregate function is a function that performs a calculation on a set of values, and returns a single value.

The most commonly used SQL aggregate functions are:

MIN() - returns the smallest value within the selected column <br>
MAX() - returns the largest value within the selected column <br>
COUNT() - returns the number of rows in a set <br>
SUM() - returns the total sum of a numerical column <br>
AVG() - returns the average value of a numerical column <br>

Aggregate functions ignore null values (except for COUNT()).


```python
payment_df_8 = pd.read_sql("""SELECT COUNT(*) as count
                              ,AVG(amount) as avg_amount
                              ,MAX(amount) as max_amount
                              FROM payment;""", connection_db)
print(payment_df_8.shape)
payment_df_8.head()
```

## Grouping Data:

The `GROUP BY` statement groups rows that have the same values into summary rows, like "find the number of customers in each country".

The `GROUP BY` statement is often used with aggregate functions (COUNT(), MAX(), MIN(), SUM(), AVG()) to group the result-set by one or more columns.

```python
payment_df_9 = pd.read_sql("""SELECT customer_id
                              , COUNT(rental_id) as total_dvd_rented
                              , SUM(amount) as total_customer_spent
                              FROM payment
                              GROUP BY customer_id
                              ORDER BY customer_id;""", connection_db)
print(payment_df_9.shape)
payment_df_9.head()
```

* GROUP BY clauses can contain multiple
columns
* Every column in your SELECT statement
must be present in a GROUP BY clause,
except for aggregated calculations
* NULLs will be grouped together if your
GROUP BY column contains NULLs

* `WHERE` does not work for groups
* `WHERE` filters on rows
* Instead use `HAVING` clause to filter for
groups

```python
payment_df_10 = pd.read_sql("""SELECT customer_id
                               , COUNT(rental_id) as total_dvd_rented
                               , SUM(amount) as total_customer_spent
                               FROM payment
                               GROUP BY customer_id
                               HAVING COUNT(rental_id)<15
                               ORDER BY customer_id;""", connection_db)
print(payment_df_10.shape)
payment_df_10
```

## CASE WHEN:
The `CASE` expression goes through conditions and returns a value when the first condition is met (like an if-then-else statement). 

* Once a condition is true, it will stop reading and return the result. 

* If no conditions are true, it returns the value in the `ELSE` clause.

* If there is no `ELSE` part and no conditions are true, it returns `NULL`.

```python
payment_df_11 = pd.read_sql("""SELECT *,
                               CASE WHEN amount > 5 THEN 'expensive'
                               ELSE 'not_expensive'
                               END as expensive_flag
                               FROM payment;""", connection_db)
print(payment_df_11.shape)
payment_df_11.head()
```

# Creating Tables:

This part is concerned with creating tables both permanent and temporary. First we will create a new database called `ba_bachelors`

```python
### Creating a new database ###

connection_ba = sqlite3.connect('ba_bachelors.db')

tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", connection_ba)
```

```python
print(tables['name'])
```

```python
## create a cursor object that allows us to interact with the database
cursor_obj = connection_ba.cursor()  
```

```python
# Creating Table:

cursor_obj.execute("DROP TABLE IF EXISTS students") # drop table if already present

students_table = """CREATE TABLE students (
                    student_id VARCHAR(255) PRIMARY KEY
                    , first_Name CHAR(25) NOT NULL
                    , last_Name CHAR(25)
                    , number_courses INT
                    , gpa DECIMAL(8,2)
                 ); """

cursor_obj.execute(students_table)

print("Table is Ready")
```

```python
students_df = pd.read_sql("""SELECT *
                             FROM students;""", connection_ba)
students_df 
```
